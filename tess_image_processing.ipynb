## Learning tesseract
### basic functions and samples

```python
"""Setup cell."""
from PIL import Image
import pytesseract
pytesseract
```

```python
bill_img = Image.open("Test-Samples/metro/metro_sample.jpg")
text = pytesseract.image_to_string(bill_img)
display(bill_img)

print(text)
```

### proccessing an image

```python
# converting to greyscale mode
grey_bill_img = bill_img.convert("L")
display(grey_bill_img)
```

```python
def binarize_img(img, treshhold_value):
    img = img.convert("L") # converting to greyscale mode
    
    width, height = img.size
    
    for x_axis in range(width):
        for y_axis in range(height):
            if img.getpixel((x_axis, y_axis)) < treshhold_value:
                img.putpixel((x_axis, y_axis), 255)
            else:
                img.putpixel((x_axis, y_axis), 0)
    
    return img

binarized_img = binarize_img(grey_bill_img, 60)
binarized_img.save("Test-Samples/binarized.jpg")
display(binarized_img.resize((400, 600), Image.ANTIALIAS)) # for better quality delete resize
```

## binarization for another samples
